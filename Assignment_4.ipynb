```python
import os
import wget
import time
import requests
import csv
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

project_path = os.path.abspath(os.path.dirname('__file__'))
driver_path = os.path.join(project_path, "chromedriver")
driver = webdriver.Chrome(driver_path)
driver.get('https://search.damai.cn/search.htm?spm=a2oeg.home.category.ditem_1.36fa23e12riVaD&ctl=%E8%AF%9D%E5%89%A7%E6%AD%8C%E5%89%A7&order=1&cty=%E4%B8%8A%E6%B5%B7')

header = ['Title', 'Venue', 'Time', 'Price']
data = []

for i in range(0,3):
    html = driver.page_source
    soup = BeautifulSoup(html)
    for item in soup.find_all("div", attrs = {"class":"items"}):
        Title = item.find ("div", attrs = {"class":"items__txt__title"}).text.strip()
        Venue = item.find ("div", attrs = {"class":"items__txt__time"}).text.strip()
        Time1 = item.find ("a", attrs = {"class":"items__txt__time__icon"})
        for el in Time1.find_parents("div",attrs = {"class":"items__txt__time"}):
            Time = el.get_text().strip()
        Price = item.find ("div", attrs = {"class":"items__txt__price"}).find("span").text.replace("\n"," ").strip()
        data.append((Title,Venue,Time,Price))
    searchBtn = driver.find_element_by_class_name('btn-next')
    searchBtn.click()
driver.quit()
df = pd.DataFrame(data, columns = header)
df.to_csv('drama.csv',sep=',', encoding = 'utf_8_sig')
df
```
